```python
#used to plot and create the desired pattern of slope (plt.plot)
```

```python
y=x*x
```

```python

plt.plot(x,y,'g*-')   
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Exponent graph')
plt.show()
```

```python

plt.plot(x,y,'g*',linestyle='dashed',linewidth=2, markersize=12)   
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Exponent graph')
plt.show()
```

```python
#plt.subplot(nrows,ncols,position)-> Used to print multiple graphs at once
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'b*-')
plt.subplot(2,2,3)
plt.plot(x,y,'g-*')
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y,'b')
plt.show()
```

```python
#Bar plot
```

```python
x=[2,8,10]
y=[11,16,9]
x1=[3,9,11]
y1=[6,5,7]
plt.bar(x,y)
plt.bar(x1,y1)
plt.title('Bar')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
#Histograms
```

```python
a=np.array([20,30,40,10,34,54,56,7,55,76,90,12,23])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

```python
#box plot using matplotlib
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True , patch_artist=True)
plt.show()
```

```python
labels ='Python','C++','HTML','JAVA'
sizes=[215,130,245,210]
colors=['Gold','Blue','Green','Pink']
explode=(0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
