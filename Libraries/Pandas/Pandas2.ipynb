```python
from io import StringIO,BytesIO
```

```python
data=('Column1,Column2,Column3\n'         #this data frame is of string type...
       'x,y,1\n'
       'a,b,3\n'
       'c,d,4\n'    
)
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))      #this is used to view dataframe as a csv file 
```

```python
df=pd.read_csv(StringIO(data) , usecols=['Column1','Column2'])   #to only specifically call or print desired column names
```

```python
df.head()
```

```python
df.to_csv('Test2.csv')  #converts the dataframe to csv and saves in directory
```
